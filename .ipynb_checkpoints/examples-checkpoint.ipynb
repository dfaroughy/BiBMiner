## Extract citations in BiBTeX format 


Give an arXiv number as input:

```python
ref=2204.02875
cite=get_bibtex(ref)
print(cite.bibtex)
```

```python
ref='hep-ph/9905221'
cite=get_bibtex(ref)
print(cite.bibtex)
```

...or a Bibtex PID generated by HEP Inspires:

```python
ref='Camargo-Molina:2022ord'
cite=get_bibtex(ref)
print(cite.bibtex)
```

HEP Inspires identifiers can also be used: 

```python
ref=2063986
cite=get_bibtex(ref)
print(cite.bibtex)
```

Finally, if the paper is not in the Inpires repository, or the id has a typo you get the following error messages:

```python
ref='ChavezFrias:666sob'
cite=get_bibtex(ref)
print(cite.bibtex)
```

Report numbers from the main Labs and experimental collaborations are supported:

```python
refs = ['CERN-TH-2002-147',
        'CERN-TH-4254/85',
        'CERN-EP-2017-100',
        'ATLAS-CONF-2021-025',
        'LHCB-PAPER-2017-013',
        'FERMILAB-PUB-15-107-T',
        'KEK-2019-40',
        'IPMU21-0074']

for ref in refs:
    cite=get_bibtex(ref)
    print(cite.bibtex)
```

As well as the identifiers of the main HEP journal:

```python
refs = ['Phys.Rev.Lett.109.160401',
        'PhysRevLett.109.160401',
        'Phys.Rev.Lett.122(2019)13.131803',
        'JHEP12(2019)065',
        'Phys.Rev.D98(2018)5,055003',
        'Eur.Phys.J.C80(2020)7.641',
        'SciPostPhys.7(2019)014',
        'Comput.Phys.Commun.276(2022)108330']

for ref in refs:
    cite=get_bibtex(ref)
    print(cite.bibtex)
```

## Generating the .bib file

The *make_bib* function takes as input a Latex *.tex* file and extracts the citations from the text. It generates as an output the corresponding .bib file in bibtex format. Notice that citations must be specified as
*\cite{...}* using the following citation identifiers: 

- arxiv number of the form *'2022.12345'*, *'hep-ph/123456'*, etc.
- the HEP inspire id which are purely numeric, of the form *'123456'*. 
- the Bibtex identifier generated by HEP inspires of the form *'Witten:123ab'*.

Notice that some articles may have multiple Bibtex PIDs, some being older and no longer in use. During extraction, all references with deprecaded PID's this will give rise to a warning that displays the 'newer' PID. The user can then replace in the tex file the older citation with this one. 

### Example:

```python
make_bib('tex_samples/example.tex')
```
