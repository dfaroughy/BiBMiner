## Extract citations in BiBTeX format 

 From HEP Inspires id:

```python
ref=2063986
cite=get_bibtex(ref)
print(cite.bibtex)
```

From arXiv numbers:

```python
ref=2204.02875
cite=get_bibtex(ref)
print(cite.bibtex)
```

```python
ref='hep-ph/9905221'
cite=get_bibtex(ref)
print(cite.bibtex)
```

From Bibtex identifiers generated by Inspires:

```python
cite=get_bibtex('Camargo-Molina:2022ord')
print(cite.bibtex)
```

Finally, if the paper is not in the Inpires repository, or the id has a typo you get the following error messages:

```python
cite=get_bibtex('ChavezFrias:666sob')
print(cite.bibtex)
```

## Generating the .bib file

The *make_bib* function  takes as input a .tex latex file and extracts and prints the citations in the text  and generates the corresponding .bib file in bibtex format. Notice that citations must be spcified as
*\cite{...}* with the following citation identifiers: 

- arxiv number of the form *'2022.12345'*, *'hep-ph/123456'*, etc.
- the HEP inspire id which are purely numeric, of the form *'123456'*. 
- the Bibtex identifier generated by HEP inspires of the form *'Witten:123ab'*.

Notice that some articles may have multiple Bibtex PIDs, both being valid. During extraction, this will give rise to a warning that displays the 'newer' PID. The user can then replace in the tex file the older citation with this one. 

### Example:

```python
make_bib('example.tex', output='bibliography.bib')
```
